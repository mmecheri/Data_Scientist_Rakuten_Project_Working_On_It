```python
# Classification Part
model.add(GlobalAveragePooling2D())
model.add(Dense(units = 1024, activation='relu'))

model.add(Dropout(rate = 0.2))
model.add(Dense(units = 512, activation='relu'))

model.add(Dropout(rate = 0.2))
model.add(Dense(units = 27, activation='softmax'))


## **Results of Tested Models**  

The table below presents the results of all the tested models, including their input size, training parameters, and performance metrics (**Accuracy & F1-score weighted**).  

| **Model**            | **Original Input Shape** | **Training Input Shape** | **Source Image Size** | **LR (Default)** | **Epochs** | **Batch Size** | **Accuracy** | **F1 Score Weighted** |
|----------------------|------------------------|-------------------------|----------------------|---------------|--------|------------|-----------|------------------|
| **InceptionResNetV2** | 299×299 | 299×299 | 500×500 | 0.001 | 40 | 64 | 0.62 | 0.61 |
| **InceptionResNetV2** | 299×299 | 224×224 | 256×256 | 0.001 | 40 | 64 | 0.58 | 0.57 |
| *DenseNet121*       | 224×224 | 224×224 | 256×256 | 0.001 | 40 | 64 | 0.61 | 0.61 |
| **Xception**         | 299×299 | 299×299 | 500×500 | 0.001 | 40 | 64 | 0.61 | 0.60 |
| **Xception**         | 299×299 | 224×224 | 256×256 | 0.001 | 40 | 64 | 0.58 | 0.56 |
| **InceptionV3**      | 299×299 | 299×299 | 500×500 | 0.001 | 40 | 64 | 0.60 | 0.59 |
| **InceptionV3**      | 224×224 | 256×256 | 256×256 | 0.001 | 40 | 64 | 0.57 | 0.56 |
| *MobileNetV2*      | 224×224 | 224×224 | 256×256 | 0.001 | 40 | 64 | 0.59 | 0.59 |
| *MobileNetV2*      | 224×224 | 224×224 | 256×256 | 0.001 | 20 | 32 | 0.59 | 0.58 |
| **VGG16**            | 224×224 | 224×224 | 256×256 | 0.001 | 40 | 64 | 0.58 | 0.57 |
| **VGG16**            | 224×224 | 224×224 | 256×256 | 0.001 | 20 | 64 | 0.58 | 0.57 |
| **VGG16**            | 224×224 | 128×128 | 256×256 | 0.001 | 10 | 32 | 0.54 | 0.53 |
| **VGG19**            | 224×224 | 224×224 | 256×256 | 0.001 | 40 | 64 | 0.57 | 0.56 |
| **NASNetMobile**     | 224×224 | 224×224 | 256×256 | 0.001 | 40 | 64 | 0.58 | 0.57 |
| **ResNet50 A REVOIR Exec Benchmak**         | 224×224 | 128×128 | 500×500 | 0.001 | 50 | 32 | 0.33 | 0.29 |
| **ResNet50 A REVOIR Exec Benchmak**         | 224×224 | 256×256 | 256×256 | 0.001 | 40 | 64 | 0.33 | 0.29 |
| **EfficientNetB7**   | 600×600 | 224×224 | 256×256 | 0.001 | 40 | 64 | 0.12 | 0.03 |

### Explanation of Table Columns

The table below describes the key input size parameters used in training the models, including their original pre-trained input shape, the actual input shape used during training, and the raw image size.

| **Concept**              | **Definition**                                                               | **Example in your case**                           |
|--------------------------|-----------------------------------------------------------------------------|---------------------------------------------------|
| **Original Input Shape** | The input size required by the pre-trained model on ImageNet. This is the size it was originally trained on. | EfficientNetB7 → 600×600, ResNet50 → 224×224 |
| **Training Input Shape** | The actual input size used for training (defined by `target_size` in `flow_from_directory`). Sometimes different from the original input shape to test performance variations. | `target_size=(224, 224)` or `target_size=(299, 299)` depending on the experiment |
| **Source Image Size**    | The raw image size before any preprocessing. This is the size of the dataset images before they are resized. | 500×500 (tested with 256×256 for performance comparison) |


####  *Original Input Shape* 
This is the input size the **pre-trained model** was originally trained on using **ImageNet**. It is recommended to keep this size for optimal performance in **transfer learning**.

| **Model**          | **Original Input Shape** |
|-------------------|------------------------|
| EfficientNetB7    | 600×600  |
| InceptionV3       | 299×299  |
| Xception         | 299×299  |
| DenseNet121      | 224×224  |
| ResNet50         | 224×224  |
| MobileNetV2      | 224×224  |
| VGG16 / VGG19    | 224×224  |

**Why is this important?**  
If using **transfer learning**, it is best to match this input size to get the most accurate results.

---

####  *Training Input Shape*
This is the **actual size of images used during training** after preprocessing. It is defined in `target_size` when using **`flow_from_directory`** in TensorFlow/Keras.

**Why can it be different from the Original Input Shape?**  
- Sometimes, a **smaller input size is used** to **reduce memory usage** and speed up training.
- Example: InceptionV3 was originally trained on **299×299**, but in your table, **224×224** was also tested to see how it affects performance.

**Example usage in `flow_from_directory`:**
```python
train_generator = datagen.flow_from_directory(
    'data/train', 
    target_size=(224, 224),  # Training Input Shape (testing 224×224 instead of 299×299)
    batch_size=32,
    class_mode='categorical'
)
```

### **Impact of Optimal Input Shape & Image Size Input vs Source Image Size**
📌 **Observations**:

1️⃣ **Models trained with their Optimal Input Shape perform better**  
   - Models trained with their **original input shape** (e.g., **InceptionResNetV2 at 299×299**, **Xception at 299×299**) achieved **higher F1 scores** than when trained with smaller input sizes.  
   - **Example:** **InceptionResNetV2 (299×299) obtained an F1-score of 0.61**, while reducing the training size to **224×224 lowered it to 0.57**.  

2️⃣ **Using a larger Source Image Size (500×500) improves performance for models with 299×299 input shape**  
   - **InceptionResNetV2, Xception, and InceptionV3 trained with 500×500 input images performed better** than when trained with **256×256 images**.  
   - **Example:** **Xception (500×500 → 299×299) achieved an F1-score of 0.60**, whereas **Xception (256×256 → 299×299) only reached 0.56**.  
   - **Explanation:** Large input sizes **preserve more image details** before resizing, reducing the information loss.  

3️⃣ **For models with a 224×224 input shape, reducing Source Image Size to 256×256 had no negative impact**  
   - **DenseNet121 and MobileNetV2 performed equally well with 256×256 images as with larger sizes.**  
   - This suggests that **as long as the image size is reasonably larger than the final training size (224×224), no major performance drop occurs**.  

---

📌 **Recommendation:**  
- **For models requiring 299×299 training input, use a larger Source Image Size (e.g., 500×500) to maintain high performance.**  
- **For models with a 224×224 training input, 256×256 Source Image Size is sufficient and does not negatively impact performance.**  
- **Always align Training Input Shape with the Optimal Input Shape to maximize performance.** 🚀
